```python
import time as t

empresa_data = {
    "Matriz": {
        "TI": {
            "Infraestrutura": {
                "Servidores": 50000,
                "Seguranca": 30000
            },
            "Desenvolvimento": {
                "Frontend": 20000,
                "Backend": 25000,
                "DevOps": 15000
            }
        },
        "Pronto-socorro": {
            "Medicos": 200000,
            "Enfermagem": 150000,
            "Equipamentos":50000,
            "Medicamentos":500000

        },
        "Financeiro": 90000
    }
}


def auditor(adf):
    def ado(*args, **kwargs):

        print("=" * 50)
        print("INICIANDO AUDITORIA DE ORÇAMENTO")
        print("=" * 50)

        inicio = t.time()

        print(f"Argumentos posicionais: {args[1:]}")
        print(f"Argumentos nomeados: {kwargs}")

        resultado = adf(*args, **kwargs)

        fim = t.time()
        tempo_execucao = fim - inicio

        print("=" * 50)
        print("AUDITORIA FINALIZADA")
        print(f"Tempo total de execução: {tempo_execucao:.6f} segundos")
        print("=" * 50)

        return resultado

    return ado

@auditor
def calcular_orcamento(dados, *departamentos_ignorados, **kwargs):

  # Obtendo taxa de câmbio dos kwargs
  taxa_cambio = kwargs.get("taxa_cambio", 1)
  moeda_destino = kwargs.get("moeda_destino", "BRL")

  def somar(dicionario):
    subtotal = 0

    for chave, valor in dicionario.items():

      # Ignora departamentos especificados
      if chave in departamentos_ignorados:
        print(f"Departamento ignorado: {chave}")
        continue

      # Se for outro dicionário, continua recursivamente
      if isinstance(valor, dict):
        subtotal += somar(valor)

      # Se for valor numérico, soma
      elif isinstance(valor, (int, float)):
        subtotal += valor

    return subtotal

  # These lines were incorrectly indented inside `somar` function.
  # They are now correctly de-indented to be part of `calcular_orcamento`.
  total = somar(dados)

  # Conversão financeira
  total_convertido = total * taxa_cambio

  print(f"\nMoeda destino: {moeda_destino}")
  print(f"Taxa de câmbio aplicada: {taxa_cambio}")

  return total_convertido

orcamento_total = calcular_orcamento(
empresa_data,
"RH",                    # Departamento ignorado
moeda_destino="BRL",
taxa_cambio=1
)

print(f"\nOrçamento total calculado: R$ {orcamento_total:,.2f}")
```
